What is the expected amount of photo-electron originating from nitrogen impurities for Spin-MCP?

```python
# Simulation parameters
eta = 1e-9 #nitrogen concentration@: 1ppb
a = 0.357e-9 #diamond lattice constant (m)
L = 2e-6 #thickness of illuminated area
D = 3e-6 #pore diameter
```

```python
import numpy as np
import matplotlib.pylab as plt

import matplotlib 
matplotlib.rc('xtick', labelsize=15) 
matplotlib.rc('ytick', labelsize=15)
```

```python
# Spin-MCP approach

demagn = np.linspace (5,15, 1000) #electro-static lens de-magnification factor
V = np.pi*L*(0.5*(D/demagn))**2 #volume of illuminated region
n = eta * (V/(a**3))

plt.figure(figsize=(8,5))
plt.plot(demagn, n, 'royalblue', linewidth=3)
plt.xlabel ('Magnification factor', fontsize=15)
plt.ylabel ('number P1 centres', fontsize=15)
plt.show()
```

```python
# Approach-1

d = 1e-9*np.linspace (10,1000, 1000) #electrode size
V = L*d*d #volume of illuminated region
n = eta*V/(a**3)

plt.figure(figsize=(8,5))
plt.plot(d*1e9, n, 'royalblue', linewidth=3)
plt.xlabel ('electrode size', fontsize=15)
plt.ylabel ('number P1 centres', fontsize=15)
plt.show()
```
