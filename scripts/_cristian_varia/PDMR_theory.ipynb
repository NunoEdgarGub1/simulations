```python
import numpy as np
import pylab as plt

import matplotlib 
matplotlib.rc('xtick', labelsize=15) 
matplotlib.rc('ytick', labelsize=15)
```

```python
# here we consider an applied voltage (but we may exceed the breakdown electric field)
mu0 = 4500 #cm^2/Vs
mu1 = 45000 #cm^2/Vs
tau0 = 100e-9 #s
tau1 = 2e-6 #s
U = 5 #V

L = np.linspace (10e-9, 5e-6, 1000)
gamma0 = tau0*mu0*1e-4*U/(L**2)
gamma1 = tau1*mu1*1e-4*U/(L**2)

plt.figure()
plt.fill_between (L*1e9, gamma1, gamma0, alpha=0.3)
plt.xlabel ('d (nm)', fontsize = 15)
plt.ylabel ('PE gain', fontsize = 15)
plt.xscale ('log')
plt.yscale ('log')
plt.show()
```

```python
# here we keep the electric field constant
mu0 = 4500 #cm^2/Vs
mu1 = 45000 #cm^2/Vs
tau0 = 100e-9 #s
tau1 = 2e-6 #s
E = 1e6 #V/cm

L = np.linspace (10e-9, 5e-6, 1000)
gamma0 = tau0*mu0*1e-4*E*1e2/(L) #gamma is the phoot-electric gain
gamma1 = tau1*mu1*1e-4*E*1e2/(L)

plt.figure()
plt.fill_between (L*1e9, gamma1, gamma0, alpha=0.3)
plt.xlabel ('d (nm)', fontsize = 15)
plt.xscale ('log')
plt.yscale ('log')
plt.show()
```

```python
# here we keep the applied voltage constant, but we make sure to never go beyond 
# the breakdown electric field, taken to be Eb = 10^6 V/cm
mu0 = 4500 #cm^2/Vs
mu1 = 45000 #cm^2/Vs
tau0 = 100e-9 #s
tau1 = 2e-6 #s
U = 5 #volts

L = np.linspace (10e-9, 5e-6, 1000)
E = U/L
ind = np.where (E>1e8)
E[ind] = 1e8
gamma0 = tau0*mu0*1e-4*E/(L)
gamma1 = tau1*mu1*1e-4*E/(L)


plt.figure()
plt.fill_between (L*1e9, gamma1, gamma0, alpha=0.3)
plt.xscale ('log')
plt.yscale ('log')
plt.show()
```

```python
# expected photo-current by reducing distance
# considering an experimental value of 1fA per NV, for L = 15um
# if we take the electric field to be constant at Eb = 3.3e4 V/cm

L = np.linspace (10e-9, 2e-6, 1000)
i = 1e-15*(15e-6/L)

plt.figure()
plt.plot (L*1e9, i*1e12, linewidth=2, color='crimson')
plt.xscale ('log')
plt.yscale ('log')
plt.ylabel ('photocurrent (pA)', fontsize= 15)
plt.xlabel ('L (nm)', fontsize= 15)
plt.show()
```
